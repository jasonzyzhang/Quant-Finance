```python
#to add more technical indicators for the overall stock market, as gauged by technical indicators using SPY:
SPY_tech_indicators = data.DataReader('SPY', 
                     start='1990-1-1', 
                     end='2020-11-20',
                     data_source='yahoo')
SPY_tech_indicators = pd.DataFrame(SPY_tech_indicators['Adj Close'])
```

```python
#to add moving averages for SPY
SPY_tech_indicators['MA5'] = SPY_tech_indicators['Adj Close'].rolling(window = 5).mean()
SPY_tech_indicators['MA10'] = SPY_tech_indicators['Adj Close'].rolling(window = 10).mean()
SPY_tech_indicators['MA50'] = SPY_tech_indicators['Adj Close'].rolling(window = 50).mean()
SPY_tech_indicators['MA100'] = SPY_tech_indicators['Adj Close'].rolling(window = 100).mean()
SPY_tech_indicators['MA200'] = SPY_tech_indicators['Adj Close'].rolling(window = 200).mean()
```

```python
#to add Releative Strength Index (RSI)
def RSI(stock, column="Adj Close", period=14):
    close = stock[column]
    delta = close.diff() 
    up, down = delta.copy(), delta.copy()

    up[up < 0] = 0
    down[down > 0] = 0
    
    roll_up = up.ewm(com=period - 1, adjust=False).mean()
    roll_down = down.ewm(com=period - 1, adjust=False).mean().abs()
    rs = roll_up / roll_down  
    rsi = 100-(100/(1+rs))
    
    stock['RSI'] = rsi

RSI(SPY_tech_indicators)
```

```python
#to add bollinger_band to the data
def bollinger_band(dataframe):
    dataframe['ma20'] = dataframe['Adj Close'].rolling(window=20).mean()
    dataframe['upper_band'] = dataframe['ma20']+\
                             (dataframe['Adj Close'].rolling(window=20).std()*2)
    dataframe['lower_band'] = dataframe['ma20']-\
                             (dataframe['Adj Close'].rolling(window=20).std()*2)

bollinger_band(SPY_tech_indicators)
```
